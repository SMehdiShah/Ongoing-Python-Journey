```python
# Defining a tuple with multiple data types
# A tuple is an immutable sequence type in Python
# If a single integer is written without a comma, it is considered an 'int'
# Adding a comma after a single element makes it a tuple

tup = (1, 2, 3, "ali", 4, 5)  # Tuple with integers and a string

# Checking the type of the tuple and printing it
print("Tuple Type:", type(tup))
print("Tuple Elements:", tup)

# Finding the length of the tuple
print("Length of Tuple:", len(tup))

# Accessing elements using indexing
print("First Element:", tup[0])
print("Second Element:", tup[1])
print("Third Element:", tup[2])
print("Fourth Element (String):", tup[3])

# Checking if an element exists within the tuple
if 3 in tup:
    print("3 is found in tuple.")
else:
    print("3 is not in tuple.")

# Demonstrating tuple slicing (Extracting elements from index 1 to 3)
tup2 = tup[1:4]
print("Sliced Tuple (1:4):", tup2)

# Iterating through the tuple using a for loop
print("Iterating through tuple:")
for item in tup:
    print(item, end=" ")
print()  # Newline for better formatting

# Demonstrating tuple concatenation (Joining two tuples)
tup3 = (10, 20, 30)
tup_combined = tup + tup3  # New tuple with elements of both
print("Concatenated Tuple:", tup_combined)

# Demonstrating nested tuples (Tuple inside another tuple)
nested_tup = (tup, tup3)
print("Nested Tuple:", nested_tup)

# Finding an index of an element
try:
    index_value = tup.index("ali")
    print("Index of 'ali':", index_value)
except ValueError:
    print("Element not found in tuple.")

# Counting occurrences of an element
tup_with_duplicates = (1, 2, 3, 4, 3, 5, 3)
count_3 = tup_with_duplicates.count(3)
print("Count of 3 in tuple:", count_3)

# Demonstrating tuple unpacking
a, b, c, d, e, f = tup  # Assigning tuple values to variables
print("Unpacked values:", a, b, c, d, e, f)

# Attempting tuple modification (will raise an error as tuples are immutable)
try:
    tup[0] = 100  # This should fail
except TypeError as e:
    print("Error:", e)
```
