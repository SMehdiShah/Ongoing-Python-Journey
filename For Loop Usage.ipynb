```python
# Python program to demonstrate the use of loops and range function with step values

# The range function generates a sequence of numbers.
# Syntax: range(start, stop, step)
# - 'start' is the initial value (inclusive).
# - 'stop' is the end value (exclusive).
# - 'step' determines the increment between values.

# ===================== PART 1: Printing Numbers from 1 to 19 =====================
print("Numbers from 1 to 19:")
for k in range(1, 20):  # Loop runs from 1 to 19
    print(k, end=" ")  # Print numbers in a single line
print("\n")  # Adding a newline for better readability


# ===================== PART 2: Printing Even Numbers from 1 to 19 =====================
print("Even numbers from 1 to 19:")
for k in range(1, 20):  # Loop runs from 1 to 19
    if k % 2 == 0:  # Check if the number is even
        print(k, end=" ")  # Print even numbers in a single line
print("\n")


# ===================== PART 3: Using range() with step to print even numbers =====================
print("Even numbers using range(2, 20, 2):")
for k in range(2, 20, 2):  # Starts from 2, goes till 18, increments by 2
    print(k, end=" ")
print("\n")


# ===================== PART 4: Printing Odd Numbers Using Step =====================
print("Odd numbers from 1 to 19 using range(1, 20, 2):")
for k in range(1, 20, 2):  # Starts from 1, goes till 19, increments by 2
    print(k, end=" ")
print("\n")


# ===================== PART 5: Printing Numbers in Reverse Order =====================
print("Numbers from 19 to 1 using reverse range:")
for k in range(19, 0, -1):  # Starts from 19, decrements by 1 till 1
    print(k, end=" ")
print("\n")


# ===================== PART 6: Printing Multiples of 5 Up to 50 =====================
print("Multiples of 5 up to 50:")
for k in range(5, 51, 5):  # Starts from 5, goes up to 50, increments by 5
    print(k, end=" ")
print("\n")


# ===================== PART 7: Nested Loop - Multiplication Table =====================
print("Multiplication table from 1 to 5:")
for i in range(1, 6):  # Outer loop for numbers 1 to 5
    for j in range(1, 11):  # Inner loop for multiplication (1 to 10)
        print(f"{i} x {j} = {i * j}", end="\t")  # Using f-string for formatted output
    print("\n")  # New line after each table


# ===================== PART 8: Skipping Multiples of 3 Using 'continue' =====================
print("Numbers from 1 to 19, skipping multiples of 3:")
for k in range(1, 20):  # Loop runs from 1 to 19
    if k % 3 == 0:  # If number is a multiple of 3, skip it
        continue  # Skip the rest of the loop body and continue with next iteration
    print(k, end=" ")
print("\n")


# ===================== PART 9: Breaking the Loop at a Certain Condition =====================
print("Numbers from 1 to 19, stopping at 13:")
for k in range(1, 20):
    if k == 13:  # If k reaches 13, break the loop
        print("\nLoop stopped at 13.")
        break
    print(k, end=" ")
print("\n")


# ===================== PART 10: Finding Prime Numbers Between 1 and 20 =====================
print("Prime numbers between 1 and 20:")
for num in range(2, 21):  # Loop through numbers from 2 to 20
    is_prime = True  # Assume the number is prime
    for divisor in range(2, int(num ** 0.5) + 1):  # Check divisibility from 2 to sqrt(num)
        if num % divisor == 0:
            is_prime = False  # If divisible, it's not a prime number
            break  # Exit inner loop early
    if is_prime:
        print(num, end=" ")
print("\n")


# ===================== PART 11: Using List Comprehension for Even Numbers =====================
print("Even numbers between 1 and 19 using list comprehension:")
even_numbers = [k for k in range(1, 20) if k % 2 == 0]
print(even_numbers)

# ===================== END OF PROGRAM =====================
print("\nProgram executed successfully!")
```
