```python
# String Slicing in Python

# A string is a sequence of characters, and we can extract a part of it (substring)
# using slicing, which follows the syntax: string[start:end].
# Note: The `end` index is exclusive, meaning slicing stops at `end - 1`.

# Defining a string variable
name = "Mehdi, Shah"

# Extracting a substring from index 0 to 4 (5 is exclusive)
print("Extracting characters from index 0 to 4:")
print(name[0:5])  # Output: 'Mehdi'

# Finding and printing the length of the string
print("\nTotal length of the string 'name':", len(name))  # Output: 11

# Working with another string
fruit = "Mango"

# Finding the length of the string 'Mango'
len1 = len(fruit)
print("\n'Mango' is a", len1, "characters long word.")  # Output: 5

# Extracting substring from index 1 to 2 (index 3 is excluded)
print("\nExtracting 'an' from 'Mango':", fruit[1:3])  # Output: 'an'

# Omitting the start index defaults it to 0
print("Extracting first 3 characters (same as [0:3]):", fruit[:3])  # Output: 'Man'

# Using negative slicing
# `fruit[0:-2]` means extracting from index 0 up to `length-2`
print("\nUsing negative slicing:")
print("Extracting 'Man' from 'Mango' using [0:-2]:", fruit[0:-2])  # Output: 'Man'
print("Extracting 'Man' using [0:len(fruit)-2]:", fruit[0:len(fruit) - 2])  # Output: 'Man'

# Extracting a substring using negative indices
# `fruit[-3:-1]` means extracting from index `(length-3)` to `(length-1)`
print("Extracting 'ng' from 'Mango' using [-3:-1]:", fruit[-3:-1])  # Output: 'ng'

# Another example of negative slicing
Ex = "Harry"
print("\nUsing negative slicing on 'Harry':")
print("Extracting 'ar' from 'Harry' using [-4:-2]:", Ex[-4:-2])  # Output: 'ar'

# Demonstrating full string slicing with different combinations
print("\nMore String Slicing Examples:")
sample_text = "Bioinformatics"

# Extracting a substring from index 4 to 10
print("Extracting 'inform' from 'Bioinformatics' [4:10]:", sample_text[4:10])  # Output: 'inform'

# Extracting from index 3 to the end
print("Extracting 'informatics' using [3:]:", sample_text[3:])  # Output: 'informatics'

# Extracting from start to index 7
print("Extracting 'Bioinfo' using [:7]:", sample_text[:7])  # Output: 'Bioinfo'

# Extracting the last 5 characters
print("Extracting last 5 characters using [-5:]:", sample_text[-5:])  # Output: 'atics'

# Extracting all characters except the last 3
print("Extracting all except last 3 using [:-3]:", sample_text[:-3])  # Output: 'Bioinformat'

# Extracting every second character from the string
print("Extracting every 2nd character using [::2]:", sample_text[::2])  # Output: 'Bionotmc'

# Reversing the string using slicing
print("Reversing the string using [::-1]:", sample_text[::-1])  # Output: 'scitamrofoniB'
```
