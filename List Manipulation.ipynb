```python
# ===================== LIST MANIPULATIONS =====================

# Creating a list of integers
l = [22, 45, 98, 1, 2, 3, 4, 10]
print("Original List:", l)  # Printing the original list
print("\n")

# ===================== APPENDING ELEMENT TO LIST =====================
# The append() method adds an element at the end of the list
l.append(88)  # Adds 88 at the end
print("After appending 88:", l)
print("\n")


# ===================== SORTING A LIST =====================
# Sorting the list in ascending order
l.sort()  # Arranges elements in ascending order (smallest to largest)
print("Sorted List (Ascending Order):", l)

# Sorting the list in descending order
l.sort(reverse=True)  # Arranges elements in descending order (largest to smallest)
print("Sorted List (Descending Order):", l)
print("\n")


# ===================== REVERSING A LIST =====================
# The reverse() method reverses the order of elements
l.reverse()  # Reverses the list (does not sort, just flips order)
print("Reversed List:", l)
print("\n")


# ===================== FINDING ELEMENTS IN A LIST =====================
# The index() method returns the index position of the first occurrence of an element
index_98 = l.index(98)  # Finds the index of element 98
print("Index of 98 in list:", index_98)

# The count() method counts how many times a specific value appears in the list
count_2 = l.count(2)  # Counts occurrences of element 2
print("Count of number 2 in list:", count_2)
print("\n")


# ===================== LIST COPYING & REFERENCE =====================
# Assigning list `l` to `m` (this creates a reference, not a new copy)
m = l  # Both `m` and `l` now refer to the same memory location
m[0] = 0  # Changing `m[0]` will also change `l`
print("After modifying `m`, list `l` also changes:", l)

# Creating an actual copy of `l` using the copy() method
m = l.copy()  # Now `m` is a separate copy
m[0] = 99  # Changing `m` does not affect `l`
print("After modifying copied list `m`, original list `l` remains unchanged:", l)
print("\n")


# ===================== INSERTING ELEMENTS INTO A LIST =====================
# The insert() method inserts an element at a specific position
l.insert(2, 55)  # Inserts 55 at index 2
print("After inserting 55 at index 2:", l)
print("\n")


# ===================== EXTENDING A LIST =====================
# The extend() method appends elements from another list at the end
m = [33, 44, 100]
l.extend(m)  # Extends `l` by adding elements from `m`
print("After extending list `l` with list `m`:", l)
print("\n")


# ===================== CONCATENATING TWO LISTS =====================
# Using the `+` operator to join two lists without modifying the original
m = [45, 55, 65]
k = l + m  # Creates a new list by combining `l` and `m`
print("Concatenated List `k` (without modifying original `l`):", k)
print("\n")
```
